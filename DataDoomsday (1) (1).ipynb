```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE, SelectFromModel
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, roc_auc_score
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Optional: standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
def evaluate_model(model, X_train, X_test, y_train, y_test, method_name):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='macro')  # multi-class
    return {
        'Method': method_name,
        'Accuracy': round(accuracy, 3),
        'Precision': round(precision, 3)
    }
```

```python
rf_baseline = RandomForestClassifier(random_state=84)
baseline_result = evaluate_model(rf_baseline, X_train, X_test, y_train, y_test, 'All Features')
```

```python
rfe_selector = RFE(estimator=LogisticRegression(max_iter=1000), n_features_to_select=10)
X_train_rfe = rfe_selector.fit_transform(X_train, y_train)
X_test_rfe = rfe_selector.transform(X_test)

rfe_result = evaluate_model(RandomForestClassifier(random_state=42), X_train_rfe, X_test_rfe, y_train, y_test, 'RFE')
```

```python
lasso = LogisticRegression(penalty='l1', solver='liblinear', max_iter=1000)
lasso_selector = SelectFromModel(lasso)
X_train_lasso = lasso_selector.fit_transform(X_train, y_train)
X_test_lasso = lasso_selector.transform(X_test)

lasso_result = evaluate_model(RandomForestClassifier(random_state=42), X_train_lasso, X_test_lasso, y_train, y_test, 'LASSO')
```

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=84)
rf_selector = SelectFromModel(rf_model)
X_train_rf = rf_selector.fit_transform(X_train, y_train)
X_test_rf = rf_selector.transform(X_test)

rf_result = evaluate_model(RandomForestClassifier(random_state=42), X_train_rf, X_test_rf, y_train, y_test, 'Random Forest Importance')
```

```python
results = pd.DataFrame([baseline_result, rfe_result, lasso_result, rf_result])
print(results)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# âœ… Step 1: Aggregate Disaster Impact and Aid per Country
aid_impact = emdat.groupby('Country').agg({
    'Total Affected': 'sum',
    'Total Deaths': 'sum',
    "Total Damage ('000 US$)": 'sum',
    "AID Contribution ('000 US$)": 'sum'
}).reset_index()

# âœ… Step 2: Clean NaNs and calculate a combined impact score
aid_impact = aid_impact.fillna(0)
aid_impact['Impact Score'] = (
    aid_impact['Total Affected'] * 0.5 +
    aid_impact['Total Deaths'] * 10 +
    aid_impact["Total Damage ('000 US$)"]
)

# âœ… Step 3: Plot Aid vs. Impact
plt.figure(figsize=(12, 7))
sns.scatterplot(
    data=aid_impact,
    x='Impact Score',
    y="AID Contribution ('000 US$)",
    hue='Country',
    legend=False,
    size='Impact Score',
    sizes=(20, 500)
)

plt.title("ðŸŒ International Aid vs. Disaster Impact by Country")
plt.xlabel("Disaster Impact Score (Affected + Deaths + Damage)")
plt.ylabel("Aid Contribution ('000 US$)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Add aid-to-impact ratio
aid_impact['Aid-to-Impact Ratio'] = aid_impact["AID Contribution ('000 US$)"] / (aid_impact['Impact Score'] + 1)

# ðŸ”» Under-supported: High impact, low aid
under_supported = aid_impact.sort_values(by='Aid-to-Impact Ratio').head(5)

# ðŸ”º Over-supported: Low impact, high aid
over_supported = aid_impact.sort_values(by='Aid-to-Impact Ratio', ascending=False).head(5)

print("ðŸ”» Under-Supported Countries:")
print(under_supported[['Country', 'Impact Score', "AID Contribution ('000 US$)", 'Aid-to-Impact Ratio']])

print("\nðŸ”º Over-Supported Countries:")
print(over_supported[['Country', 'Impact Score', "AID Contribution ('000 US$)", 'Aid-to-Impact Ratio']])
```

1) After comparing disaster frequencies between 2000â€“2010 and 2011â€“2024, what does the shift suggest about the impact of climate change on natural disasters?

```python
import pandas as pd

# âœ… Load EM-DAT data
emdat = pd.read_excel('/content/public_emdat_custom_request_2025-02-14_7e6ca5c3-c7ec-4190-b6c8-ff8ea20cd69e.xlsx')

# âœ… Ensure 'Start Year' is numeric
emdat['Start Year'] = pd.to_numeric(emdat['Start Year'], errors='coerce')

# âœ… Filter relevant years
period_1 = emdat[(emdat['Start Year'] >= 2000) & (emdat['Start Year'] <= 2010)]
period_2 = emdat[(emdat['Start Year'] >= 2011) & (emdat['Start Year'] <= 2024)]

# âœ… Count disasters by type
freq_1 = period_1['Disaster Type'].value_counts().rename('2000â€“2010')
freq_2 = period_2['Disaster Type'].value_counts().rename('2011â€“2024')

# âœ… Merge and calculate % change
comparison = pd.concat([freq_1, freq_2], axis=1).fillna(0)
comparison['% Change'] = ((comparison['2011â€“2024'] - comparison['2000â€“2010']) / (comparison['2000â€“2010'] + 1)) * 100  # +1 to avoid division by 0

# âœ… Round and sort by % change
comparison = comparison.round(1).sort_values(by='% Change', ascending=False)

# âœ… Display table
comparison.reset_index().rename(columns={'index': 'Disaster Type'})
```

```python
import matplotlib.pyplot as plt

comparison_plot = comparison.sort_values(by='% Change', ascending=False)

comparison_plot[['2000â€“2010', '2011â€“2024']].plot(
    kind='bar',
    figsize=(12, 6),
    title='Disaster Frequency Comparison: 2000â€“2010 vs 2011â€“2024'
)
plt.ylabel('Number of Disasters')
plt.xlabel('Disaster Type')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

2) For countries showing a steady increase in disaster frequency or INFORM risk scores, what climate-related factors might be contributing to this rise (e.g., glacier melt, urban heat islands, deforestation)?

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np

# Load EM-DAT data
emdat = pd.read_excel('/content/public_emdat_custom_request_2025-02-14_7e6ca5c3-c7ec-4190-b6c8-ff8ea20cd69e.xlsx')
emdat['Start Year'] = pd.to_numeric(emdat['Start Year'], errors='coerce')
emdat = emdat[emdat['Start Year'].between(2000, 2024)]

# Load INFORM data
inform = pd.read_excel('/content/inform_risk_mid_2024_v068.xlsx',
                       sheet_name='INFORM Risk 2024 (a-z)',
                       skiprows=2)
inform = inform.rename(columns={
    '(a-z)': 'Country',
    '(0-10)': 'INFORM Risk',
    '(0-10).17': 'Vulnerability Index',
    '(0-10).31': 'Institutional Capacity'
})
```

```python
# Disaster count per year per country
trend_df = emdat.groupby(['Country', 'Start Year']).size().reset_index(name='Disaster Count')

# Linear regression slope = disaster trend
trend_results = []
for country in trend_df['Country'].unique():
    sub_df = trend_df[trend_df['Country'] == country]
    if len(sub_df) >= 5:
        X = sub_df['Start Year'].values.reshape(-1, 1)
        y = sub_df['Disaster Count'].values
        slope = LinearRegression().fit(X, y).coef_[0]
        trend_results.append({'Country': country, 'Disaster Trend': slope})

disaster_trend_df = pd.DataFrame(trend_results)
```

```python
# Merge trend with INFORM
merged = pd.merge(disaster_trend_df, inform, on='Country', how='inner')

# Filter: steady increase + high risk
hotspots = merged[(merged['Disaster Trend'] > 0.05) & (merged['INFORM Risk'] >= 6.5)]
hotspots = hotspots.sort_values(by='Disaster Trend', ascending=False).reset_index(drop=True)

# Show top hotspots
hotspots[['Country', 'Disaster Trend', 'INFORM Risk', 'Vulnerability Index', 'Institutional Capacity']]
```

```python
def early_warning():
    clear_output()
    display(Markdown("""
âœ… **Good Call!**
Early warning messages are sent via radio + SMS.
Community leaders mobilize families to nearby shelters.

ðŸ§ª Because of your **low readiness**, the shelters are:
- Not well supplied
- Overcrowded
But at least people are **safe** from the first wave.

What do you do next?
"""))

    btn1 = widgets.Button(description="Deploy Rescue Boats Preemptively")
    btn2 = widgets.Button(description="Send Police to Enforce Evacuations")

    btn1.on_click(lambda b: boats_outcome())
    btn2.on_click(lambda b: conflict_outcome())

    display(btn1, btn2)


def no_early_warning():
    clear_output()
    display(Markdown("""
âŒ **Critical Delay**
Waiting for official confirmation cost precious hours.
By the time alerts go out, storm surge is already breaching riverbanks.

âš ï¸ Impact:
- Mass panic
- Severe casualties
- Breakdown in evacuation routes

**Outcome**: Over 10,000 displaced, hundreds missing.

ðŸ“ Lesson: High-risk countries must act fast, even with imperfect info.
"""))


def boats_outcome():
    clear_output()
    display(Markdown("""
ðŸš¤ Rescue boats launched early navigate narrow rivers.
They successfully extract:
- 80% of stranded villagers
- Deliver first aid kits and water

ðŸ‘‘ **Outcome**: Your proactive response saves lives despite low infrastructure.

ðŸ§  **Lesson**: Low readiness can be overcome with timely, community-focused action.
"""))


def conflict_outcome():
    clear_output()
    display(Markdown("""
ðŸš“ Police arrive but tensions rise.
Some communities resist, fearing theft or loss of livestock.

ðŸŽ­ Outcome:
- Delays in evacuation
- 2 rescue workers injured
- Media backlash

ðŸ§  **Lesson**: Social trust is key in disaster settings â€” readiness is not just equipment, it's communication.
"""))
```
